```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#Importing the dataset
df = pd.read_csv('titanic.csv')

#Getting to know the dataset better
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
#Checking the presence of null values in dataset
print("Here, I will be checking if any null values are present in the dataset.")
a = df.isnull().sum()
print(a, "\n")

#Visualising the null values present in dataset with the help of a heatmap
print("On visualising, the null or missing values present in the data set using heatmap.\n")
sns.heatmap(df.isnull(), yticklabels=False, cbar=False)
print("We can see that as the column 'Embarked' has 2 null values only, that's why, it is not visible in the heatmap.")
```

```python
#Cleaning the Dataset
sns.boxplot(x='Pclass', y='Age', data=df)
sns.set_style("whitegrid")

print("""Here, the mean age for different Passenger-classes(Pclass) is different. Therefore, to make the most accurate
analysis of the dataset instead of replacing the null values of age with the mean age obtained in the above describe
function, I will replace the null values of age with the respective mean ages classified according to the Pclass of the Passengers.
According to boxplot I have plotted below, the mean age of Pclass==1 passengers is around 37, the mean age of Pclass==2 passengers 
around 29 and the mean age of Pclass==3 is 24.""")

#Function to replace all the Nan values in Age
def fill_NaN_Age(cols):
  Age = cols[0]
  Pclass = cols[1]

  if pd.isnull(Age):
    if Pclass==1:
      return 37
    elif Pclass==2:
       return 29
    else:
       return 24
  else:   
    return Age   
```

```python
#Applying the above function to replace all NaN values in Age
df['Age'] = df[['Age', 'Pclass']].apply(fill_NaN_Age, axis=1)

#Also, as 'Cabin' column is not much neccessary in the analysis and doesn't provide any important info, therefore I dropped 'Cabin' column
df.drop('Cabin', axis=1, inplace=True)

#From above, as I can see 2 NaN values 'Embarked' are also present, so I will drop those 2 NaN values from the dataset.
df.dropna(inplace=True)

#Now through heatmap again on visualising I see all the NaN values of Age has been replaced, the Cabin column has been dropped and the Nan values from 'Embarked' column has also been removed. 
z = sns.heatmap(df.isnull(), yticklabels=False, cbar=False)
plt.show()
df.isnull().sum()
print("As, observed from below no missing/NaN values are present in the dataset.")
```

In the data set, as I go on analysing further I can see that the columns 'Name', 'PassengerId', and 'Ticket' cannot be divided into categories and don't contribute much to the outcome. So, as these columns are of no use, I'm dropping these columns from the dataset.

```python
df.drop(['PassengerId', 'Name', 'Ticket'], axis=1, inplace=True)
```

Here, below I changed the male and female in the 'Sex' column into 0 and 1 values respectively and stored it in the new column 'Sexnum'. After that as the 'Sex' column in the dataset is of no use as I converted it into a new column 'Sexnum', I removed 'Sex' column.

```python
def new_sex(df, col):
    def sex(x):
        if x=='female':
            return 1
        return 0
    
    return df[col].apply(sex)
    
df['SexNum'] = new_sex(df, 'Sex')
df.drop('Sex', axis=1, inplace=True)
```

```python
# Graphical Analysis
print(df['Survived'].value_counts(), "\n")


sns.countplot(x='Survived', data=df, palette=["#FF0000", "#028C6A"])
sns.set_theme()
plt.plot()  
```

***Conclusion 1:- Here in above plot, we can see '0' stands for passengers who died and '1' stands for passengers who survived. A total of 549 passsengers died and a total of 340 passengers survived. Hence, I conclude that the overall passengers survival ratio is '340 : 549' = '0.62 : 1'***

```python
sns.countplot(x='Survived', hue='SexNum', data=df, palette=['#1f19d2','#ff3366'])
sns.set_style("whitegrid")
```

***For reference:- Sexnum{0:'Male', 1:'Female'} and Survived{0:'Died', 1:'Survived'})***

***Conclusion 2:- Here, by observing the plot above I conclude that number of deaths of male passengers is more than the number of deaths of female passengers, which means the rescue of female passengers was given more priority.***

```python
palette = ["#FF0000", "#4CC417"]

df[df['Pclass'] == 1].Survived.groupby(df.Survived).count().plot(kind='pie', figsize=(12,5), autopct='%1.1f%%', colors=palette)
plt.axis('equal')
plt.legend(["Died","Survived"])
plt.title("Survival Rate of First Class Passengers", fontsize=19, color= 'blue')
plt.show()

print("\n")
df[df['Pclass'] == 2].Survived.groupby(df.Survived).count().plot(kind='pie', figsize=(12,5), autopct='%1.1f%%', colors=palette)
plt.axis('equal')
plt.legend(["Died","Survived"])
plt.title("Survival Rate of Second Class Passengers", fontsize=19, color= 'blue')
plt.show()

print("\n")
df[df['Pclass'] == 3].Survived.groupby(df.Survived).count().plot(kind='pie', figsize=(12,5), autopct='%1.1f%%', colors=palette)
plt.axis('equal')
plt.legend(["Died","Survived"])
plt.title("Survival Rate of Third Class Passengers", fontsize=19, color= 'blue')
plt.show()
```

***(For reference- Survived{0:Died, 1:Survived})***

***Conclusion 3:- On observing the above 3 pie-charts all together, I conclude that the survival rate of First class passengers is the highest being 63.0%, followed by the Second class passenger being 47.3%, followed by the Third class passenger being 24.2%.***
