# Feature Engineering

```python
#Splitting the target variable
target_var = df['SalePrice']
target_var.head()
```

```python
#Visualising the distribution of Dependent variable- 'SalePrice'
sns.distplot(target_var, hist=True)
```

***From above plot, I conclude that the distribution of Saleprice is left skewed. So, to make it normally distributed, I will be using log transformation.***

```python
target_log = np.log(target_var)
```

```python
sns.distplot(target_log, hist=True)
```

```python
plt.rcParams['figure.figsize'] = (19.0,13.0)
prices_var = pd.DataFrame({"SalePrice":df['SalePrice'], "Log_SalePrice":target_log})
prices_var.hist()
```

***Hence, after log transformation we can see that the distribution of the dependent variable- SalePrice is normalised.***

```python
#Dropping the target variable from dataset
raw_data = df
df = df.drop(['SalePrice'], axis=1)
df.head()
```

```python
#MSSubClass: The building class
#Changing MSSubClass into a categorical variable
df['MSSubClass'] = df['MSSubClass'].astype(str)
```

```python
#OverallCond: Overall condition rating
#Changing OverallCond into a categorical variable
df['OverallCond'] = df['MSSubClass'].astype(str)
```

```python
"""MoSold: Month Sold
YrSold: Year Sold
The month and year in which house is sold are converted into a categorical variable"""

df['MoSold'] = df['MoSold'].astype(str)
df['YrSold'] = df['YrSold'].astype(str)
```

```python
"""TotalBsmtSF: Total square feet of basement area
1stFlrSF: First Floor square feet
2ndFlrSF: Second floor square feet"""

#Adding all the square feet features 
df['total_SF'] = df['TotalBsmtSF'] + df['1stFlrSF'] + df['2ndFlrSF']

#I'm here removing the TotalBsmtSF, 1stFlrSF, 2ndFlrSF
df.drop(['TotalBsmtSF', '1stFlrSF', '2ndFlrSF'], axis=1)
df.head()
```

```python
#Saving all the categorical columns in a list
categorical_cols = [col for col in df.columns.values if df[col].dtype == 'object'] 

#Dataframe with categorical features
df_cat = df[categorical_cols]

#Dataframe with numerical features
df_num = df.drop(categorical_cols, axis=1)
```

***Here, I'm reducing the skewness of the numeric features***

```python
df_num.hist(figsize=(16,20), bins=50, xlabelsize=7, ylabelsize=7)
```

```python
from scipy.stats import skew
df_num_skew = df_num.apply(lambda x:skew(x.dropna()))
df_num_skew = df_num_skew[df_num_skew > .75]

#Here, I'm applying +1 log transformation for all numerical features with skewness over .75 
df_num[df_num_skew.index] = np.log1p(df_num[df_num_skew.index])
```

```python
df_num_skew
```

```python
df_num.hist(figsize=(17,19), bins=50, xlabelsize=7, ylabelsize=7)
```

# Model Building

#### REGRESSION

1)LINEAR REGRESSION

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(new_df, target_log, test_size = 0.20, random_state=0)
print("X_train", X_train.shape, "\n", "X_test", X_test.shape, "\n", "y_train", y_train.shape, "\n", "y_test", y_test.shape)
```

***Linear Regression Base Model***

```python
import statsmodels.api as sm
model1 = sm.OLS(y_train, X_train).fit()
```

```python
#Here, I'm printing the Linear Regression Summary
model1.summary()
```

```python
def rmse(prediction, target):
  differences = prediction - target
  differences_squared = differences**2
  meanof_differences_squared = differences_squared.mean()
  rmse_val = np.sqrt(meanof_differences_squared)
  return rmse_val
```

```python
columns1 = ['Model', 'R-squared score', 'Adj.R-squared score', 'RMSE']
report_of_model1 = pd.DataFrame(columns = columns1)
predictions1 = model1.predict(X_test)
```

```python
temp1 = pd.Series({'Model':'Linear Regression Model','R-squared score':model1.rsquared,
                   'Adj.R-squared score':model1.rsquared_adj, 'RMSE':rmse(predictions1, y_test)})
model1_finalreport = report_of_model1.append(temp1, ignore_index=True)
model1_finalreport
```
