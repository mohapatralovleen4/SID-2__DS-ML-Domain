1) Write python code to calculate Mean, Median, Mode, Variance and Standard Deviation of the numbers in the below list without using any libraries.

```python
numbers = [488, 296, 374, 189, 326, 229,  14, 162, 917, 778, 908, 376, 567,
162, 923, 130, 263, 655, 577, 486, 961, 966, 792,  19, 941, 722,
286, 191, 138, 704, 481, 230, 437, 592, 244, 885,   7, 524, 514,
483,  89, 779, 279, 478, 185, 940, 385,  43, 163, 571, 867, 492,
269,  23, 796, 888, 898, 725, 337, 573, 375, 916,  90, 743, 117,
742, 944, 358, 547, 975, 499, 469, 721, 665, 639,  70, 612, 338,
733, 634, 358, 524, 598, 466, 696, 691, 511, 450, 699,  76, 216,
359, 859, 754, 210, 309, 745,  38, 934, 541]

# write your code here

#Mean of the list 'numbers'
sum_of_entries = sum(numbers)
num_of_entries = len(numbers)
mean = (sum_of_entries/num_of_entries)
print("The mean of list \"numbers\" is: ", mean )

#Median of the list 'numbers'
#Here, I am sorting the list 'numbers'
numbers.sort()
#Here, I am finding the position of the median
if len(numbers) % 2 == 0:
   first_median = numbers[len(numbers) // 2]
   second_median = numbers[len(numbers) // 2 - 1]
   median = (first_median + second_median) / 2
else:
   median = numbers[len(numbers) // 2]
print("The median of list \"numbers\" is: ", median)

#Mode of the list 'numbers'
import collections
#Here, I have calculated the frequency of each item
info = collections.Counter(numbers)
info_list = dict(info)
#Here, I am calculating the highest frequency
max_value = max(list(info.values()))
mode = [num for num, freq in info_list.items() if freq == max_value]
if len(mode) == len(numbers):
   print("No mode in the list")
else:
   print("The mode of the list \"numbers\" is : " + ', '.join(map(str, mode)))

#Variance of the list 'numbers'
mean = sum(numbers) / len(numbers)
res = sum((i - mean) ** 2 for i in numbers) / len(numbers)
print("The variance of list \"numbers\" is : ", res)

#Standard Deviation of the list 'numbers'
mean = sum(numbers) / len(numbers)
variance = sum([((x - mean) ** 2) for x in numbers]) / len(numbers)
res = variance ** 0.5
print("The standard deviation of list \"numbers\" is : ", res)
```

2) Use the lambda and map() to square the numbers in the numbers list and store them in a newly created list known as squared. Then use filter() to create a  new list of all even numbers from the squared list.

```python
# write your code here

#Here, I am using lambda to find the square of numbers in the list 'numbers'
squared = list(map(lambda n : n**2, numbers))

#Here, I am filtering out the list of sqaured numbers into a new list containing only even squared numbers 
even_numbers_squared = list(filter(lambda s : s%2 == 0, squared)) 

print("The list of squared numbers from the list 'numbers' is:\n ", squared)
print("\nThe list of squared even numbers filtered out from the list 'squared' is:\n ", even_numbers_squared)
```

6)Find the number of males in the age range of 30 to 60 (including both 30 and 60 here), who have the occupation of "Self_Employed"

```python
# write your code here

cc_df = pd.read_csv("creditcard.csv")

#Here, I have filtered out the dataframe 'cc_df' into new dataframe 'filtered_df' containing the columns "Gender", "Age" and "Occupation" only
filtered_df = cc_df.filter(items = ["Gender", "Age", "Occupation"])

#Here, I have filtered out the previous dataframe into a new dataframe 'x' containing Column "Gender" with only 'Male' Datas only
x = filtered_df[filtered_df['Gender'] == 'Male']

#Here, I have filtered out the previous dataframe into a new dataframe 'y' containing Column "Gender" with 'Male' and "Occupation" with "Self_Employed" Datas only
y = x[x["Occupation"] == 'Self_Employed']

#Finally, I filtered out the previous dataframe into a new dataframe 'final_df' containing data about Males who are Self_Employed in the age range of 30 to 60(including both 30 and 60)
z = y[ 30 <= y["Age"]]
final_df = z[z["Age"] <= 60]

#Here, the count of the number of males in the age range of 30 to 60(including both 30 and 60), who have the occupation of "Self_Employed"
count_male = sum(final_df.value_counts())
print("The number of males in the age range of 30 to 60(including both 30 and 60), who have the occupation of Self_Employed is:", count_male)
```

7)Find the number of males and females who were active in the last 3 months (if the Is_Active column shows "Yes", that means that customer was active in the last 3 months, else he/she wasn't).

```python
# write your code here

#Here, I filtered out the dataframe 'cc_df' into a new dataframe'new_df' consisting of Columns "Gender" and "Is_Active" only
new_df = cc_df.filter(items = ["Gender", "Is_Active"])

#Here, I filtered out the previous dataframe 'new_df' into a new dataframe 'new_active' consisting of people active in the last 3 months
new_active = new_df[new_df["Is_Active"] == 'Yes']

#Here, finally I filtered out the 'new_active' dataframe and made a new dataframe 'male_df' consisting of males active in the last 3 months 
male_df = new_active[new_active["Gender"] == 'Male']

#Here, finally I filtered out the 'new_active' dataframe and made a new dataframe 'female_df' consisting of females active in the last 3 months
female_df = new_active[new_active["Gender"] == 'Female']

print("The number of males who were active in the last 3 months is:", sum(male_df.value_counts()))
print("\nThe number of females who were active in the last 3 months is:", sum(female_df.value_counts()))
```

8)Find the percentage of null values in the column "Credit_Product", and replace them all with the value which is lesser among "yes" or "no" in the column.

```python
# write your code here

null_percentage = (cc_df['Credit_Product'].isnull().sum() / cc_df['Credit_Product'].shape[0]) * 100
print("The percentage of null values in the column 'Credit_Product' is:", null_percentage, "\n")

cc_df["Credit_Product"].value_counts()
print("As the number of 'yes' is lesser among 'yes' or 'no', I have replaced all the null values in the 'Credit_Product' column with the 'Yes' value")

cc_df['Credit_Product'].fillna('Yes', inplace = True)
cc_df['Credit_Product']
```
